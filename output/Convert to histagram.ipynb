```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

```python
file = "tc1o.pgm"
text_file = open(file)
y = []
x = []
for i in range(1,255):
    t = text_file.readline()
    t = t.replace(":",",")
    t = t.replace("answer"," ")
    t = list(map(int, t.split(",")))
    y.append(t[0])
    x.append(t[1])
# print(his)
figure(figsize=(20, 8), dpi=80)
plt.bar(y,x)
plt.title("Pixel Histogram")
plt.xlabel("Pixel")
plt.ylabel("Count")
```
